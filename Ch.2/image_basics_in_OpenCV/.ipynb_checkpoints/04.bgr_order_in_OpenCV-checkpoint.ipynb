**We already mentioned that OpenCV uses the BGR color format instead of the RGB one. This can be seen in the following diagram, where you can see the order of the three channels:**



```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
print('Done!')
```

```python
# Read in the image with cv2, in the BGR format
img_OpenCV = cv2.imread('logo.png')

# Display with matplotlib
plt.imshow(img_OpenCV)
plt.show()
```

```python
# Split the BBR image into its unique channels
(B,G,R) = cv2.split(img_OpenCV)
```

```python
# Merger the cahannels into the RGB format supported by Matplotlib
matplot_img = cv2.merge([R,G,B])
```

#### Plotting both images with matplotlib subplots

```python
plt.figure(figsize=(6, 10))
plt.subplot(121)
plt.title('BGR Format of OpenCV')
plt.imshow(img_OpenCV)
plt.subplot(122)
plt.title('RGB Format of Matplotlib')
plt.imshow(matplot_img)
plt.show()
```

Clearly we can see the effect of the different channel formats. BGR of the first image excatly loads the Blue, then The Green and then the Red channels, while RGB of the second image loads the Red, then the Green and then the Blue channels.

#### Showing both images using cv2.imshow()

**Showing them in separate windows**

* Show both images (img_OpenCV and img_matplotlib) using cv2.imshow()
* This will show the image in true color:
* cv2.imshow('bgr image', img_OpenCV)
* This will show the image in wrong color:
* cv2.imshow('rgb image', img_matplotlib)
* cv2.waitKey(0)
* cv2.destroyAllWindows()

```python
cv2.imshow('BGR Image Format Right', img_OpenCV)
cv2.imshow('RGB Image Format Wrong', matplot_img)
cv2.waitKey(0)
```

Interestingly cv2 displays the BGR image with RGB channels anti-clockwise in the right order, and displays the RGB image with BGR channels anti-clockwise in the wrong order.

**Showing them in same window**

if we want to show the two images in the same window, we can build a full image that contains the two images, concatenating them horizontally. To do so, we will use NumPy's `concatenate()` method. The parameters of this method are the two images to concatenate and the axis. In this case, `axis = 1 (to stack them horizontally)`:

```python
# To stack horizontally (img_OpenCV to the left of img_matplotlib):
img_concats = np.concatenate((img_OpenCV, matplot_img), axis=1)
# Now, we show the concatenated image:
cv2.imshow('BGR Image and RGB image', img_concats)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Using Numpy for Split.

cv2.split() is time consuming and slow. A better option is to slice the images with Numpy


* Get the three channels (instead of using cv2.split):
* B = img_OpenCV[:, :, 0]
* G = img_OpenCV[:, :, 1]
* R = img_OpenCV[:, :, 2]

```python
B, G, R = img_OpenCV[:,:,0], img_OpenCV[:,:,1], img_OpenCV[:,:,2]
np.max(B)
```

#### Converting images from BGR to RGB in one line of numpy code...

```python
plt.imshow(img_OpenCV)
plt.show()
```

Let's convert this image from BGR to RGB in one line of Numpy code...

```python
# First let's see how to reverse a simple list in python
x = [1, 2, 3, 4]

# reverse x
x[::-1]
```

```python
# Now we reverse the channel order of a 3D numpy array

img_OpenCV_to_RGB = img_OpenCV[:,:,::-1]
plt.imshow(img_OpenCV_to_RGB)
plt.show()
```

#### Reversing a Gray Scale Image

```python
gray_img = cv2.imread('logo.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(gray_img)
plt.show()
```

```python
# Let's reverse the gray Image
gray_img_reversed = gray_img[:,::-1]
plt.imshow(gray_img_reversed)
plt.show()
```

**Reversing the gray scale image literally flips it sideways to the opposite direction**
